```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as ani
from PIL import Image
```

```python
M = 64  # Ilosc obrazów
m = np.linspace(-M/2,M/2,M)
def image_generator(IS,IO):
    theta = np.linspace(0, 2*np.pi, 100) # Kąty od 0 do 2*pi
    k = 1
    for i in m:
        r =  np.sin(IS*theta+((i*np.pi)/10))
    
        # Utworzenie wykresu biegunowego
        dpi = 1024
        figsize = (1, 1)  # 1 cal * 256 dpi = 256 pikseli
        fig = plt.figure(figsize=figsize, dpi=dpi)
        ax = fig.add_subplot(111, polar=True)
        ax.plot(theta, r)
        plt.xticks([])  # usuwa ticki osi X
        plt.yticks([])  # usuwa ticki osi Y
        # Dodanie tytułu
       # ax.set_title("Śmigło o {n} łopatach w {m} momencie".format(n = n, m = round(i,2)))
        saveFormat = "img/{IS}/{k}.png".format(IS=IS,k=k)
        k += 1
        plt.savefig(saveFormat)
        plt.close(fig)
image_generator(3,M)
image_generator(5,M)
```

```python
def aliasing(IS):
    fullimage = Image.new("RGB",(1024,1024))
    i = 1
    chunk = 16
    for L in range(64):
        temp_img = Image.open("img/{IS}/{x}.png".format(IS=IS,x=i))
        image_crop = temp_img.crop((0,L*chunk,1024,(L+1)*chunk))
        fullimage.paste(image_crop,(0,L*chunk))
        i+=1
    fullimage.show()
aliasing(3)
aliasing(5)
```
