```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = 64  # Ilosc obrazw
n = 3 # Ilosc smigie
m = np.linspace(-M/2,M/2,M)
theta = np.linspace(0, 2*np.pi, 100) # Kty od 0 do 2*pi
r =  np.sin(5*theta+((m[0]*np.pi)/10)) # Wzr na funkcj (np. spirala)

# Utworzenie wykresu biegunowego
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(theta, r)

# Dodanie tytuu
ax.set_title("Przykad wykresu biegunowego")

# Wywietlenie wykresu
plt.show()
```
